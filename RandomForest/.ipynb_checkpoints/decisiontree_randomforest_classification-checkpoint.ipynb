여기서는 엑셀파일로 준비된 데이터를 이용해서 decision tree 및 이의 ensemble method 중 하나인 random forest를 시험해 봅니다. 
엑셀파일에서 데이터는 예제 xlsx파일을 참고하여 첫 번째 세로줄이 label, 두 번째 세로줄부터 feature가 되도록 준비하면 됩니다. 
label은 0,1,2,3, ... 의 숫자로 준비하는 것이 좋습니다. 


```python
%matplotlib inline

import pandas
import xlrd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

아래의 네 줄의 코드는 decision tree를 시각화 하는 데에 필요한 부분인데, 
각자의 환경에 따라 코드를 다르게 수정해야 될 수 있습니다.
decision tree를 시각화할 필요가 없으면 실행하지 않아도 됩니다.

# Data Preparation
여기서는 엑셀파일로 미리 준비된  training set과 test set을 사용합니다.

```python
data0 = pandas.read_excel('breastcancer_train.xlsx')
y_vals_train = data0.values[:,0]
x_vals_train = data0.values[:,1:]
data1 = pandas.read_excel('breastcancer_test.xlsx')
y_vals_test = data1.values[:,0]
x_vals_test = data1.values[:,1:]
```

```python
data0
```

# Decision Tree
Decision tree는 아래 두 줄로 생성할 수 있습니다.
random state를 0으로 한 것은 이 예제에서 생성되는 3개의 방법에 대한 비교를 공평하게 하기 위함입니다.
코드를 실행시켜 보면 class_weight, criterion, max_depth, max_features, max_leaf_nodes 등 조절할 수 있는 파라미터가 꽤 많음을 알 수 있습니다. 여기서 가장 중요한 파라미터를 꼽으라면 max_depth를 들 수 있을 것 같습니다. 가지치는 횟수를 미리 제한하는 것으로 이것만으로도 overfitting이 꽤 줄어든다고 합니다. 

```python
tree = DecisionTreeClassifier(random_state=0)
tree.fit(x_vals_train, y_vals_train)
```

다음은 생성된 TREE를 시각화하는 과정으로 PDF파일로도 저장하게 하였습니다.
이를 위한 준비가 제대로 되지 않은 경우 에러가 날 수 있습니다. 그 경우 그냥 스킵하시면 됩니다.

```python
export_graphviz(tree, out_file = "tree.dot", class_names=["malignant","benign"],feature_names=data0.columns[1:])
with open("tree.dot") as f:
    dot_graph = f.read()
    dot=graphviz.Source(dot_graph)
    dot.render('tree')
display(graphviz.Source(dot_graph))
```

```python
print("accuracy on training set: {:.3f}".format(tree.score(x_vals_train, y_vals_train)))
print("accuracy on test set: {:.3f}".format(tree.score(x_vals_test, y_vals_test)))
```

가지치기를 끝까지 진행하였기 때문에 training set에 대한 정확도는 무조건 1이 나오게 되어 있으며 이는 무의미합니다.
test set에 대한 정확도가 training set과 큰 차이가 없어야 좋은 model이 만들어 졌다고 볼 수 있습니다. 아래와 같이 max_depth를 4로 제한하여 다시 decision tree를 만들어 봅니다.

```python
tree4 = DecisionTreeClassifier(max_depth=4,random_state=0)
tree4.fit(x_vals_train, y_vals_train)
```

```python
export_graphviz(tree4, out_file = "tree4.dot", class_names=["malignant","benign"],feature_names=data0.columns[1:])
with open("tree4.dot") as f:
    dot_graph = f.read()
    dot=graphviz.Source(dot_graph)
    dot.render('tree_maxdepth_4')
display(graphviz.Source(dot_graph))
```

```python
print("accuracy on training set: {:.3f}".format(tree4.score(x_vals_train, y_vals_train)))
print("accuracy on test set: {:.3f}".format(tree4.score(x_vals_test, y_vals_test)))
```

max_depth를 조절함에 따라 test set에 대한 accuracy가 달라지는 것을 확인할 수 있을 것입니다. 
decision tree의 장점은 시각화의 용이하면서 feature의 중요도를 어느 정도 알려줄 수 있습니다. 
아래의 코드는 각 특성별 중요도를 나타내는 과정으로 모든 중요도의 합은 1입니다.

```python
n_features = x_vals_train.shape[1]
plt.barh(range(n_features), tree4.feature_importances_,align='center')
plt.yticks(np.arange(n_features), data0.columns[1:])
plt.xlabel("importance")
plt.ylabel("feature")
plt.ylim(-1, n_features)
```

```python
tree4.feature_importances_
```

# Random Forest
Decision tree로부터 계산된 특성중요도는 아마도 몇몇 feature만 높은 값을 가지게 될 것입니다.
또한, decision tree는 max_depth를 제한하더라도 여전히 overfitting의 가능성이 높은 편이라고 합니다. 
이를 극복하기 위해 여러 개의 tree를 만들어 이로부터 결과를 예측하는 방법이 제안되었는데,
그 중 대표적인 방법인 random forest를 이용해 보겠습니다.
random forest의 중요한 파라미터는 tree를 몇 개를 만드냐 입니다. 
간단히 n_estimator의 숫자를 바꾸어 주면 됩니다.
대체로 n_estimator의 숫자가 커질수록 성능이 향상되지만 계속 올라가지는 않습니다. 
또한, 너무 숫자가 크게되면 연산시간이 매우 길어집니다.
decision tree와 마찬가지로 max_depth를 바꾸어 주는 것도 결과에 어느 정도 영향을 줍니다만,
대개 'auto'로 되어 있는 default값을 그대로 사용한다고 합니다. 

```python
forest = RandomForestClassifier(n_estimators=1000, random_state=0)
forest.fit(x_vals_train, y_vals_train)
```

```python
print("accuracy on training set: {:.3f}".format(forest.score(x_vals_train, y_vals_train)))
print("accuracy on test set: {:.3f}".format(forest.score(x_vals_test, y_vals_test)))
```

```python
n_features = x_vals_train.shape[1]
plt.barh(range(n_features), forest.feature_importances_,align='center')
plt.yticks(np.arange(n_features), data0.columns[1:])
plt.xlabel("importance")
plt.ylabel("feature")
plt.ylim(-1, n_features)
```

하나의 decisin tree만 사용하였을 때보다 대부분의 경우 test set에 대한 accuracy가 올라가고, 보다 많은 feature들에서 높은 특성 중요도 값을 보여줄 것 입니다.
